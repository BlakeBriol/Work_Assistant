```python
#Methods
import pandas as pd
import _strptime
import csv
from datetime import datetime, date
from sys import exit
import os

#For inputting work entries
def data_input():
    # Raw data inputs from user
    resp = input("Is this for today? (n/y)\n")
    if resp == "y":
        now = datetime.today()
        fdate = now.strftime('%Y-%m-%d')
    elif resp == "n":
        fdate = input("What was/will be the work date? (YYYY-MM-DD)\n")
    else:
        print("Please enter 'y' or 'n'. Please try again.")
        return  # Exit the function instead of using exit()

    #Asks user for inputs
    respA = input("What Job was this for?\n")
    respB = input("What time did you clock in? HH:MM \n")
    respC = input("What time did you clock out? HH:MM \n")
    respD = input("What rate?\n")
    
    # Converts time responses into datetime values
    t1 = datetime.strptime(respB, "%H:%M")
    t2 = datetime.strptime(respC, "%H:%M")
    
    # Finds amount of time worked in terms of hours
    Hworked = t2 - t1
    Tworked = (Hworked.total_seconds()) / 3600
    
    # Float value for rate response
    Rate = float(respD)
    
    # Earnings calculation
    Earnings = Tworked * Rate
    
    # Converts responses into DataFrame
    data = {
        'Job': [respA],
        'Date': [fdate],
        'Clocked In': [respB],
        'Clocked Out': [respC],
        'Duration': [Tworked],
        'Hourly Rate': [Rate],
        'Earnings': [Earnings]
    }
    
    df_new = pd.DataFrame(data)
    
    # Check if the CSV file actually exists lmaoo
    try:
        df_existing = pd.read_csv("Hours.csv")
        # Append the new entry
        required_headers = ['Job','Date','Clocked In','Clocked Out','Duration','Hourly Rate','Earnings']
        actual_headers = df_existing.columns.tolist()
        if actual_headers == required_headers:
            df_combined = pd.concat([df_existing, df_new], ignore_index=True) #Concat is annoying to use, bring back append!
            df_combined.to_csv("Hours.csv", index=False)
            
        else:
            
            df_NH.columns = required_headers
            df_NH.to_csv('Hours.csv', index=False)
            
            df_combined = pd.concat([df_NH, df_new], ignore_index=True) #Concat is annoying to use, bring back append!
            df_combined.to_csv("Hours.csv", index=False)
            
    except FileNotFoundError:
        # Creates the file if there isnt one
        df_new.to_csv("Hours.csv", index=False)
    
    print("Data appended.")

#This just prints the dataframe
def see_df():
    print(pd.read_csv("Hours.csv"))


#For Clearing All Data
def clear_data():
    filename = "Hours.csv"
    
    #opening the file with w+ mode truncates the file
    f = open(filename, "w+")
    f.close()


#Gives total earnings
def give_total():
    filename = pd.read_csv("Hours.csv")
    print(filename["Earnings"].sum())


#Gives actual total earnings lmaoo
def taxed_total():
    filename = pd.read_csv("Hours.csv")
    tax_total = filename["Earnings"].sum()*0.75
    print(tax_total)


#Need I say more?
def Agraph():
    filename = pd.read_csv("Hours.csv")
    filename.plot()


#Gives back some statistical information
def avg_csv():
    filename = pd.read_csv("Hours.csv")
    
    #Gives Averages of earnings, duration, and hourly rate
    print("Averages\n")
    print(f"Earnings = {filename['Earnings'].mean()}\n")
    print(f"Duration = {filename['Duration'].mean()}\n")
    print(f"Hourly Rate = {filename['Hourly Rate'].mean()}\n")
    
    #Gives Modes of job and clock times
    print("Modes\n")
    print(f"Clock In Time = {filename['Clocked In'].mode().iloc[0]}\n")
    print(f"Clock Out Time = {filename['Clocked Out'].mode().iloc[0]}\n")
    print(f"Job = {filename['Job'].mode().iloc[0]}\n")


#Clears a row given by user
def clear_row():
    n = input("What is the index of the row you would like to clear?")
    b = int(n)
    with open("Hours.csv", 'r') as file:
        reader = csv.reader(file)
        rows = list(reader)

    if 0 <= b < len(rows):
        rows.pop(b)
    else:
        print("Index Not Found")
        return

    with open("Hours.csv", 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(rows)
    
def backup():
    if os.path.exists("Hours(BACKUP).csv"):
        os.remove("Hours(BACKUP).csv")  # Remove the existing file to ensure a new file
    
    #Extracts data from Hours.csv
    with open("Hours.csv", 'r') as Ofile:
        stuff = Ofile.read()

    # Puts data from Ofile into Nfile
    with open("Hours(BACKUP).csv", 'w') as Nfile:
        Nfile.write(stuff)


```

```python
#Main Function
def HourTracker():
    Q = input("""What would you like to do?\n
    1 for dataframe print\n
    2 for data input\n
    3 for total revenue\n
    4 for taxed total\n
    5 for a fun graph\n
    6 for averages\n
    7 to create a non-live backup\n
    8 for row clear\n
    9 for data clear\n
    10 for exit\n
    """)


    #Dateframe print
    if Q == "1":
        I1 = see_df()
        print(f"\n\n\n{I1}")

    #Data Input
    elif Q == "2":
        backup()
        data_input()
        #print(f"\n\n\n {I2}")

    #Total Rvenue
    elif Q == "3":
        I3 = give_total()
        print(f"\n\n\n {I3}")

    #Taxed Revenue
    elif Q == "4":
        I4 = taxed_total()
        print(f"\n\n\n {I4}")

    #Graph
    elif Q == "5":
        I5 = Agraph()
        print(f"\n\n\n {I5}")


    #Averages
    elif Q == "6":
        I6 = avg_csv()
        print(f"\n\n\n{I6}")

    #Backup 
    elif Q == "7":
        backup()
        print("Backup Created!")

    #Row Clear
    elif Q == "8":
        clear_row()
    
    #Data Clear
    elif Q == "9":
        Q2 = input("Are you sure?(y/n)")
        if Q2 == "y":
            Q3 = input("Okay, but actually? (n/y)")
            if Q3 == "y":
                clear_data()
            else:
                print("Okay")
        else:
            print("Okay")

    #To exit the program
    elif Q == "10":
        print("Exit Successful")

    #For unsupported input
    else:
        print("Unsupported Operation")

    
```
